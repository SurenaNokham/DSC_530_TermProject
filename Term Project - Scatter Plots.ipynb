# Create two scatter plots comparing two variables and provide your analysis on correlation and causation. Remember, covariance, Pearsonâ€™s correlation, and Non-Linear Relationships should also be considered during your analysis (Chapter 7).

```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
outbreaks_data = pd.read_csv('FoodborneOutbreaks.csv')

# Remove rows with missing values
outbreaks_data = outbreaks_data[['Illnesses', 'Hospitalizations']].dropna() 
```

```python
# Select the variables of interest
illnesses = outbreaks_data['Illnesses']
hospitalizations = outbreaks_data['Hospitalizations']

# Create Scatter plot
plt.scatter(illnesses, hospitalizations)
plt.xlabel('Reported Illnesses')
plt.ylabel('Hospitalizations')
plt.title('Scatter Plot of Illnesses vs Hospitalizations')
plt.show()

# Calculate Correlation analysis
correlation = np.corrcoef(illnesses, hospitalizations)[0,1]
print('Correlation coefficient:', correlation)

# Causation analysis
covariance = np.cov(illnesses, hospitalizations)[0,1]
print('Covariance:', covariance)

# Pearson's correlation
pearson_corr = illnesses.corr(hospitalizations)
print("Pearson's correlation coefficient:", pearson_corr)

# Non-linear relationship analysis
sns.regplot(x=illnesses, y=hospitalizations, scatter_kws={'alpha':0.5, 'color':'blue'})
plt.xlabel('Reported Illnesses')
plt.ylabel('Hospitalizations')
plt.title('Scatter Plot with Regression Line')
plt.show()
```

```python
# Correlation: 
# The scatter plot shows a positive linear relationship between the "Illnesses" and "Hospitalizations"
# variables. As the number of illnesses increases, hospitalizations will increase too. The correlation
# coefficient is 0.50 which indicates a strong correlation (strong range is between 0.50-1.0).

# Causation:
#The high correlation between "Illnesses" and "Hospitalizations" suggests that there is a relationship between
# the two variables. However, it doesn't necessarily mean that one variable is causing the other. Other factors
# could be contributing to both variables. Further analysis is needed to establish a causal relationship.

# Covariance:
# A positive, high covariance (275.72) suggests a positive relationship between the two variables but it does not
# tell us the strength of the relationship.

# Pearsons Correlation:
# With a value of 0.50, the positive coefficient supports a positive linear relationship between both variables.
```
