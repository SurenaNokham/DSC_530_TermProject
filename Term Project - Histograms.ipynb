```python
# Import only the necessary modules from matplotlib
import matplotlib.pyplot as plt

# Import pandas, numpy and seaborn with their commonly used aliases
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Read csv file and assign data to variable
outbreaks_data = pd.read_csv('FoodborneOutbreaks.csv')
```

```python
# Calculate the frequency of outbreaks by month
outbreaks_by_month = outbreaks_data['Month'].value_counts().head(12)

# Plot the frequency using Pandas plotting function
outbreaks_by_month.plot.bar()
plt.title('Outbreak Case Frequency by Month (2005-2015)')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.show()

# No outliers oberved.
```

```python
# Plot the histogram
plt.hist(outbreaks_data['Year'], bins=10, color='pink', edgecolor='black')

# Set the title, x-axis label and y-axis label
plt.title('Confirmed Cases of Outbreak by Year')
plt.xlabel('Year')
plt.ylabel('Confirmed Cases')

#Display the plot          
plt.show()

# No outliers observed.
```

```python
# Calculate which states had the most outbreaks
topstate = outbreaks_data['State'].value_counts().head(6)

# # Plot the frequency 
plt.bar(topstate.index, topstate.values, color='orange', edgecolor='black')
plt.title('Top States with the Highest Frequency of Outbreak')
plt.xlabel('State')
plt.ylabel('Frequency')
plt.show()

# No outliers observed.
```

```python
# Calculate contaminated ingredients
top_ingredient = outbreaks_data['Ingredient'].value_counts().head(6)

# Set title, x-axis label, and y-axis label
plt.bar(top_ingredient.index, top_ingredient.values, color='green')
plt.title('Top 5 Contaminated Ingredients in Outbreaks')
plt.xlabel('Ingredient')
plt.ylabel('Frequency')

# Display histogram
plt.show()

# No outliers observed.
```

```python
# Calculate the frequency of outbreaks by genus
genus_data = outbreaks_data['Genus'].value_counts().head(10)

# Plot the frequency 
genus_data.plot.bar(color='purple')
plt.title('Most Common Genus Types in Foodborne Illnesses')
plt.xlabel('Genus')
plt.ylabel('Frequency')
plt.show()
```
