# Plot 1 analytical distribution and provide your analysis on how it applies to the dataset you have chosen (Chapter 5).

```python
# Import libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Load the dataset
outbreaks_data = pd.read_csv('FoodborneOutbreaks.csv')
```

```python
# Create a mapping dictionary for months
month_mapping = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12}

# Convert the 'Month' column to integers using the mapping dictionary
outbreaks_data['Month'] = outbreaks_data['Month'].map(month_mapping)
```

```python
# Month columns converted to numeric values and NaN are dropped
months = pd.to_numeric(outbreaks_data['Month'], errors='coerce').dropna().values

# Calculate mean and std deviation
mean = np.mean(months)
std_dev = np.std(months)

# Random 1000 samples generated
normal_dist = np.random.normal(mean, std_dev, 1000)

# Plot histogram
plt.hist(normal_dist, bins=30, density=True, alpha=0.5, color='yellow', edgecolor='black')
plt.xlabel('Month')
plt.ylabel('Probability Density')
plt.title('Normal Distribution of Months')
plt.show()
```

```python
# Looking at the generated normal distribution and comparing it the Outbreak dataset, we can get an idea of when 
# outbreaks will occur for the different months. The normal distribution peaks between 5.0-7.5 which coalates
# to the following months, respectively: May, June, July, mid-August. With it peaking in the summer months, there 
# is a high probability of outbreaks to occur. This distribution is evenly spread, indicating a relatively 
# consistent occurrence of outbreaks throughout the year.
```
