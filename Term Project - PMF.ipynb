```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

outbreaks_data = pd.read_csv('FoodborneOutbreaks.csv')
```

```python
# Create Scenario 1: Outbreaks in California
# Create Scenario 2: Outbreaks in all other states

scenario1 = 'California'
scenario2 = 'Other States'     
```

```python
# Create PMF for each scenario

pmf_scenario1 = outbreaks_data.loc[outbreaks_data['State'] == scenario1, 'State'].value_counts(normalize=True)
pmf_scenario2 = outbreaks_data.loc[outbreaks_data['State'] != scenario1, 'State'].value_counts(normalize=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
pmf_scenario1.plot(kind='bar')
plt.title('PMF of Outbreaks in ' + scenario1)
plt.xlabel('Number of Outbreaks')
plt.ylabel('Probability')

plt.subplot(2, 1, 2)
pmf_scenario2.plot(kind='bar')
plt.title('PMF of Outbreaks in ' + scenario2)
plt.xlabel('Number of Outbreaks')
plt.ylabel('Probability')

plt.show()
```
